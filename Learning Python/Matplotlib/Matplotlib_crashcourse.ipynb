```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style
```

```python
#Scatter graph
X = np.random.random(50) *100
y = np.random.random(50) *100
plt.scatter(X,y, c='brown', marker='*', s=150)
```

```python
#Line graph
year = [2003 + x for x in range(10)]
weight = [3, 4, 5, 12, 20, 30, 35, 42, 48, 51]
plt.plot(year, weight, c='orange', lw=8)
```

```python
#Bar graph
names = ['Alex', 'Bob', 'Charlie', 'David', 'Elijah']
eatenchocolates = [13, 14, 66, 43, 100]

plt.bar(names, eatenchocolates, color='blue', align='edge', width=0.5, edgecolor='y')
```

```python
#Histogram

agedistri = np.random.normal(20, 3, 1000) #Normal distribution of ages with mean 20 yrs, std of 3 and sample size of 1000
versiona = plt.hist(agedistri, bins=20)
plt.show()
versionb = plt.hist(agedistri, bins=[agedistri.min(), 17, 23, agedistri.max()])
plt.show()
```

```python
versionc = plt.hist(agedistri, bins=20, cumulative=True)
```

```python
#Pie chars
langs = ['python', 'C#', 'C++', 'Java']
num =  [100, 23, 25, 10]
explodes = [0, 0, 0.3, 0]

plt.pie(num, labels=langs, explode=explodes, autopct='%.2f%%', startangle=90)
plt.show()
```

```python
years = np.array([2017 + x for x in range(8)])
income = np.random.randint(1000, 3000, size=(8))
income_ticks = np.arange(500, 3500, 250)


plt.plot(years, income)
plt.title('Income changes throughout the years(In SGD)', fontsize=20, fontname='Calibri')
plt.xlabel('Years', fontsize=15)
plt.ylabel('Income($)', fontsize=15)
plt.yticks(income_ticks, [f"${x}"for x in income_ticks])
```

```python

a = np.random.randint(10, 50, size=(8))
b = np.random.randint(20, 60, size=(8))
c = np.random.randint(30,70,size=(8))   

plt.plot(years, a, label='companyA')
plt.plot(years, b, label='companyB')
plt.plot(years, c, label='companyC')
yticks=np.arange(0,75,5)
plt.yticks(yticks)
plt.legend(loc='best')
```

```python
#Multiple figures

x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2, y2)

```

```python
#Subplots

figs, axs = plt.subplots(1,2)

axs[0].plot(np.arange(100), np.sin(np.arange(100)))
axs[0].set_title('Sin graph') #Use set_title instead  of just title
axs[0].set_xlabel('try') #use set_xlabel instead of just xlabel
axs[1].scatter(np.random.random(100), np.random.random(100))
axs[1].set_title('Random Scatter graph')

plt.suptitle('2 graphs') #Title of all graphs in this instance
#Exporting plots
plt.savefig('2plots.jpg', dpi=300, transparent=True) #Saves the figure as a file in the same folder. Add dpi=int to change quality
```

```python
#3D plotting
x = np.random.random(100)
z= np.random.random(100)
y = np.random.random(100)

plt.axes(projection='3d')
plt.scatter(x,y,z)
plt.title('3d graph')
plt.show()
```

```python
#Surfaceplot

ax = plt.axes(projection='3d')

x = np.arange(-5, 5)
y= np.arange(-5,5)

X, Y = np.meshgrid(x,y)

z =np.sin(X)*np.sin(Y)
ax.plot_surface(X,Y,z, cmap='Spectral') #IMPORTANT
ax.view_init(azim=0, elev=10) #azim meaning azimuth, elev meaning elevation
plt.show()
```

```python
#Making animations. Suggest you copy paste this code to a .py file instead
rng = np.random.default_rng()
heads_tails = [0,0]

for i in range(10):
    heads_tails[rng.integers(0,2)] += 1
    plt.bar(["Heads", "Tails"], heads_tails, color=["Orange", "Yellow"])
    plt.pause(0.01)
plt.show()
```
